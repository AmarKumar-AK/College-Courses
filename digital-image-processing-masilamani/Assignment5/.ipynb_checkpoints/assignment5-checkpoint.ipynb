# Q.1

```python
from cmath import exp, pi
 
def fft(x):
    N = len(x)
    if(N <= 1):
        return x
    
    even = fft(x[0::2])
    odd =  fft(x[1::2])
    
    temp = [exp(-2j*pi*k/N)*odd[k] for k in range(N//2)]
    
    return [even[k] + temp[k] for k in range(N//2)] + [even[k] - temp[k] for k in range(N//2)]

def ifft(x):
    N = len(x)
    if(N <= 1):
        return x
    
    even = ifft(x[0::2])
    odd =  ifft(x[1::2])
    
    temp = [(1/N)*(exp(2j*pi*k/N)*odd[k]) for k in range(N//2)]
    
    return [even[k] + temp[k] for k in range(N//2)] + [even[k] - temp[k] for k in range(N//2)]
```
