```python
# reading image
img = cv2.imread("baboon.png")
# converting image from BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# converting image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# plotting images
plt.subplot(1,2,1)
plt.title("grayscale image")
plt.imshow(gray,cmap="gray")

plt.subplot(1,2,2)
plt.title("original image")
plt.imshow(img)
```

```python
# defining filters
filter1 = [[1,1,1],[0,0,0],[-1,-1,-1]]
filter2 = [[-1,0,1],[-1,0,1],[-1,0,1]]
```

```python
# function to take convolution of image and filter
def convolution(img,filters):
    # matrix having all elements as zero
    new_matrix = np.zeros((img.shape[0]-2,img.shape[1]-2))
    
    for i in range(0,img.shape[0]-2):
        for j in range(0,img.shape[1]-2):
            new_matrix[i][j] = (img[i][j] * filters[0][0]) + (img[i][j+1] * filters[0][1]) + (img[i][j+2] * filters[0][2]) + (img[i+1][j] * filters[1][0]) + (img[i+1][j+1] * filters[1][1]) + (img[i+1][j+2] * filters[1][2]) + (img[i+2][j] * filters[2][0]) + (img[i+2][j+1] * filters[2][1]) + (img[i+2][j+2] * filters[2][2])
    print(new_matrix)
```

```python
convolution(gray,filter1)
```

```python
convolution(gray,filter2)
```
