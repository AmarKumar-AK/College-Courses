```python
import numpy as np
import matplotlib.pyplot as plt
feat_mat = [[1,6],[7,2],[8,9],[9,9],[2,1],[2,2],[2,4],[7,1]]
y = [[0],[0],[0],[0],[1],[1],[1],[1]]
```

```python
X=np.array(feat_mat)
y=np.array(y)
y=y.reshape(-1,1)
```

```python
plt.scatter(X[:,0],X[:,1])
print(y.shape)
```

```python
X=np.append(np.ones(y.shape),X,1)
for i in range(0,y.shape[0]):
    if(y[i][0])==0:
        X[i][0]=X[i][0]*-1
        X[i][1]=X[i][1]*-1
        X[i][2]*=-1
```

```python
plt.scatter(X[:,0],X[:,1])
print(X)
b=[[1],[1],[1],[1],[1],[1],[1],[1]]
b = np.array(b)
```

```python
def plotter(W):
    if W[2][0] == 0 :
        print("0 Division attempted skipping iteration")
        return
    x_var = np.linspace(0,9,100)
    y_pts= (-W[0][0] - W[1][0]*x_var)/W[2][0]
    plt.figure()    
    plt.plot(x_var,y_pts)
    X_disp=np.array(feat_mat)
    plt.scatter(X_disp[0:4,0],X_disp[0:4,1],c="red")
    plt.scatter(X_disp[4:8,0],X_disp[4:8,1],c="green")
    plt.title("equation " + str(W[1][0])+"x + "+str(W[2][0])+"y + "+str(W[0][0]))
```

```python
z = np.linalg.pinv(X)
print(z.dot(b))
plotter(z.dot(b))
```
