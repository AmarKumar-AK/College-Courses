```python
import numpy as np
import matplotlib.pyplot as plt
feat_mat = [[2,2],[-1,-3],[-1,2],[0,-1],[1,3],[-1,-2],[1,-2],[-1,-1]]
y = [[1],[0],[1],[0],[1],[0],[0],[1]]
```

```python
X=np.array(feat_mat)
y=np.array(y)
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
X=np.append(np.ones(y.shape),X,1)
for i in range(0,y.shape[0]):
    if(y[i][0])==0:
        X[i][0]=X[i][0]*-1
        X[i][1]=X[i][1]*-1
        X[i][2]*=-1
```

```python
plt.scatter(X[:,0],X[:,1])
print(X)
```

```python
def plotter(W):
    if W[2][0] == 0 :
        print("0 Division attempted skipping iteration")
        return
    x_var = np.linspace(-2,2,100)
    y_pts= (-W[0][0] - W[1][0]*x_var)/W[2][0]
    plt.figure()    
    plt.plot(x_var,y_pts)
    X_disp=np.array(feat_mat)
    plt.scatter(X_disp[:,0],X_disp[:,1])
    plt.title("at W as "+str(W))
```

```python
W = [[0],[0],[0]]
learning_rate = 0.1
theta = [[0.05],[0.05],[0.05]]
criterion = [[100],[1000],[1000]]
l = 0
while np.linalg.norm(theta)<np.linalg.norm(criterion):
    miss_class_sum = np.zeros([3,1])
    missclassified = []
    l=l+1
    print(l)
    for i,each in enumerate(X) :
        dot_prd = np.dot(each,W)
        if dot_prd <= 0 :
            missclassified.append(i)
    if len(missclassified)==0 :
        print("done")
        break
    for each in missclassified:
        a=X[each,:]
        for i in range(0,len(a)):
            miss_class_sum[i] = miss_class_sum[i] + a[i]
    criterion = learning_rate * miss_class_sum
    W = W + criterion
    plotter(W)
```
