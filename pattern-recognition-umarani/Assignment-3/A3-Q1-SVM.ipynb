```python
import numpy as np
import matplotlib.pyplot as plt
feat_mat = [[0,0],[1,0],[0,1],[1,1]]
y = [[0],[0],[0],[1]]
```

```python
X=np.array(feat_mat)
y=np.array(y)
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
X=np.append(np.ones(y.shape),X,1)
for i in range(0,y.shape[0]):
    if(y[i][0])==0:
        X[i][0]=X[i][0]*-1
        X[i][1]=X[i][1]*-1
        X[i][2]*=-1
```

```python
plt.scatter(X[:,0],X[:,1])
print(X)
```

```python
from sklearn.linear_model import SGDClassifier
clf = SGDClassifier(loss = "hinge" , penalty ="l2",max_iter = 1000)
tmp=np.array(feat_mat)
clf.fit(tmp,y)
W = clf.coef_
w = clf.coef_[0]
a = -w[0] / w[1]
xx = np.linspace(-5, 5)
yy = a * xx - (clf.intercept_[0]) / w[1]
plt.figure()
plt.plot(xx, yy, 'k-')
plt.scatter(tmp[:,0],tmp[:,1])
```
